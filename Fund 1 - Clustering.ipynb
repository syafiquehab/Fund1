```python
klcired_clusterer = cluster.KMeans(init = 'k-means++', n_clusters=2)
krkmeans = klcired_clusterer.fit_predict(klci_red)
```

```python
import numpy as np
rainbow  = np.array(['red', 'green', 'blue', 'orange', 'purple', 'teal', 'magenta', 'cyan'])
```

```python
plt.scatter(klci_red[:,0], klci_red[:,1], color = rainbow[krkmeans])
plt.scatter(klcired_clusterer.cluster_centers_[:,0], klcired_clusterer.cluster_centers_[:,1], marker='+', s=30)
plt.title("KMeans on FBMKLCI and Redemption")
plt.xlabel("FBMKLCI (scaled)")
plt.ylabel("Redemption (scaled)")
```

```python
krcluster_centers = {}
krcluster_membership = {}
for cluster_count in range (2, 15):
    krkm = cluster.KMeans(init='k-means++', n_clusters=cluster_count)
    krcluster_membership[cluster_count] = krkm.fit_predict(klci_red)
    krcluster_centers[cluster_count] = krkm.cluster_centers_
```

```python
import sklearn.metrics as metrics
kr = {}
for cluster_count in range (2,15):
    kr[cluster_count] = metrics.silhouette_score(klci_red, krcluster_membership[cluster_count])
    print cluster_count, kr[cluster_count]
```

# k-means and DBScan clustering on Fund 1 Sales and Redemption

```python
sales_red = fund1[["Sales", "Redemption"]]
srscaler = StandardScaler().fit(sales_red)
sales_red = srscaler.transform(sales_red)
plt.scatter(sales_red[:,0], sales_red[:,1])
plt.xlabel("Sales")
plt.ylabel("Redemption")
plt.title("Sales vs Redemption")
```

```python
salesred_clusterer = cluster.KMeans(init = 'k-means++', n_clusters=2)
srkmeans = salesred_clusterer.fit_predict(sales_red)
plt.scatter(sales_red[:,0], sales_red[:,1], color = rainbow[srkmeans])
plt.scatter(salesred_clusterer.cluster_centers_[:,0], salesred_clusterer.cluster_centers_[:,1], marker='+', s=30)
plt.title("KMeans on Sales and Redemption")
plt.xlabel("Sales (scaled)")
plt.ylabel("Redemption (scaled)")
```

```python
srcluster_centers = {}
srcluster_membership = {}
for cluster_count in range (2, 15):
    srkm = cluster.KMeans(init='k-means++', n_clusters=cluster_count)
    srcluster_membership[cluster_count] = srkm.fit_predict(sales_red)
    srcluster_centers[cluster_count] = srkm.cluster_centers_

sr = {}
for cluster_count in range (2,15):
    sr[cluster_count] = metrics.silhouette_score(sales_red, srcluster_membership[cluster_count])
    print cluster_count, sr[cluster_count]
```

```python
plt.scatter(sr.keys(), sr.values())
```

```python
srdbsc = cluster.DBSCAN(eps = 0.5, min_samples = 15).fit(sales_red)
srlabels = srdbsc.labels_
srcore_samples = np.zeros_like(srlabels, dtype = bool)
srcore_samples[srdbsc.core_sample_indices_] = True
unique_srlabels = np.unique(srlabels)
colors = plt.cm.Spectral(np.linspace(0,1, len(unique_srlabels)))
for (label, color) in zip(unique_srlabels, colors):
    class_member_mask = (srlabels == label)
    xy = sales_red[class_member_mask & srcore_samples]
    plt.plot(xy[:,0],xy[:,1], 'o', markerfacecolor = color, markersize = 5)
    
    xy2 = sales_red[class_member_mask & ~srcore_samples]
    plt.plot(xy2[:,0],xy2[:,1], 'o', markerfacecolor = color, markersize = 5)
plt.title("DBSCAN on Sales and Redemption")
plt.xlabel("Sales (scaled)")
plt.ylabel("Redemption (scaled)")
```

# k-means and DBScan clustering on KLCI and Fund 1 Sales

```python
klci_sales = fund1[["FBMKLCI", "Sales"]]
ksscaler = StandardScaler().fit(klci_sales)
klci_sales = ksscaler.transform(klci_sales)
plt.scatter(klci_sales[:,0], klci_sales[:,1])
plt.xlabel("FBMKLCI")
plt.ylabel("Sales")
plt.title("FBMKLCI vs Sales")
```

```python
klcisales_clusterer = cluster.KMeans(init = 'k-means++', n_clusters=2)
kskmeans = klcisales_clusterer.fit_predict(klci_sales)
plt.scatter(klci_sales[:,0], klci_sales[:,1], color = rainbow[kskmeans])
plt.scatter(klcisales_clusterer.cluster_centers_[:,0], klcisales_clusterer.cluster_centers_[:,1], marker='+', s=30)
plt.title("KMeans on FBMKLCI and Sales")
plt.xlabel("FBMKLCI (scaled)")
plt.ylabel("Sales (scaled)")
```

```python
kscluster_centers = {}
kscluster_membership = {}
for cluster_count in range (2, 15):
    kskm = cluster.KMeans(init='k-means++', n_clusters=cluster_count)
    kscluster_membership[cluster_count] = kskm.fit_predict(klci_sales)
    kscluster_centers[cluster_count] = kskm.cluster_centers_

ks = {}
for cluster_count in range (2,15):
    ks[cluster_count] = metrics.silhouette_score(klci_sales, kscluster_membership[cluster_count])
    print cluster_count, ks[cluster_count]
```
