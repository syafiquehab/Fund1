```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
```

```python
fund = pd.read_csv('Fund1.csv', parse_dates=[[0,1]], index_col='Month_Year', thousands=',').abs()
fund
```

```python
fund1 = fund.rename(columns={'Income Distributed' : 'Income_Distributed', 'Fund Size': 'Fund_Size'})
```

# k-means and DBScan clustering on FBMKLCI and Fund 1 Redemption

```python
# redemption_klci = fund1.drop(fund1.columns[[0,2,3,4,5]], axis=1).abs()
klci_red = fund1[["FBMKLCI", "Redemption"]]
klci_red
```

```python
# redemption_klci['Redemption_Growth'] = redemption_klci.Redemption / redemption_klci.Redemption.shift(1)-1
# redemption_klci['FBMKLCI_Return'] = redemption_klci.FBMKLCI / redemption_klci.FBMKLCI.shift(1)-1
# redemption_klci
```

```python
# plt.scatter(klci_red.FBMKLCI, klci_red.Redemption)
```

```python
from sklearn.preprocessing import StandardScaler
krscaler = StandardScaler().fit(klci_red)
klci_red = krscaler.transform(klci_red)
```

```python
plt.scatter(klci_red[:,0], klci_red[:,1])
plt.xlabel("FBMKCLI")
plt.ylabel("Redemption")
plt.title("FBMKLCI vs Redemption")
```

```python
import sklearn.cluster as cluster
```

```python
# klci_with_value = redemption_klci[redemption_klci.FBMKLCI_Return.notnull()]
```

```python
# interesting_columns = fund1[['FBMKLCI', 'Redemption']]
```

```python
klcired_clusterer = cluster.KMeans(init = 'k-means++', n_clusters=2)
krkmeans = klcired_clusterer.fit_predict(klci_red)
```

```python
import numpy as np
rainbow  = np.array(['red', 'green', 'blue', 'orange', 'purple', 'teal', 'magenta', 'cyan'])
```

```python
plt.scatter(klci_red[:,0], klci_red[:,1], color = rainbow[krkmeans])
plt.scatter(klcired_clusterer.cluster_centers_[:,0], klcired_clusterer.cluster_centers_[:,1], marker='+', s=30)
plt.title("KMeans on FBMKLCI and Redemption")
plt.xlabel("FBMKLCI (scaled)")
plt.ylabel("Redemption (scaled)")
```

```python
krcluster_centers = {}
krcluster_membership = {}
for cluster_count in range (2, 15):
    krkm = cluster.KMeans(init='k-means++', n_clusters=cluster_count)
    krcluster_membership[cluster_count] = krkm.fit_predict(klci_red)
    krcluster_centers[cluster_count] = krkm.cluster_centers_
```

```python
import sklearn.metrics as metrics
kr = {}
for cluster_count in range (2,15):
    kr[cluster_count] = metrics.silhouette_score(klci_red, krcluster_membership[cluster_count])
    print cluster_count, kr[cluster_count]
```

```python
plt.scatter(kr.keys(), kr.values())
```

```python
krdbsc = cluster.DBSCAN(eps = 0.5, min_samples = 3).fit(klci_red)
```

```python
krcore_samples_mask = np.zeros_like(krdbsc.labels_, dtype = bool)
krcore_samples_mask[krdbsc.core_sample_indices_] = True
krlabels = krdbsc.labels_
krn_clusters_ = len(set(krlabels)) - (1 if -1 in krlabels else 0)
print('Estimated number of clusters: %d' % krn_clusters_)
print("Silhouette Coefficient: %0.3f" % metrics.silhouette_score(klci_red, krlabels))
```

```python
unique_krlabels = set(krlabels)
colors = plt.cm.Spectral(np.linspace(0,1, len(unique_krlabels)))
for k, col in zip(unique_krlabels, colors):
    if k == -1:
        # Black used for noise.
        col = 'k'

    krclass_member_mask = (krlabels == k)

    krxy = klci_red[krclass_member_mask & krcore_samples_mask]
    plt.plot(krxy[:, 0], krxy[:, 1], 'o', markerfacecolor=col,
             markeredgecolor='k', markersize=14)

    krxy = klci_red[krclass_member_mask & ~krcore_samples_mask]
    plt.plot(krxy[:, 0], krxy[:, 1], 'o', markerfacecolor=col,
             markeredgecolor='k', markersize=6)

plt.title("DBSCAN on FBMKLCI and Redemption")
plt.xlabel("FBMKLCI (scaled)")
plt.ylabel("Redemption (scaled)")
```
