```python
from matplotlib import pyplot
import pandas as pd
import numpy as np
```

```python
fund = pd.read_csv('new_fund1.csv', thousands=',')
fund1 = fund.rename(columns={'Income Distributed' : 'Income_Distributed', 'Fund Size': 'Fund_Size'})
fund1
```

```python
new_fund1 = fund1[fund1.Month == "Dec"]
new_fund1
```

```python
pyplot.plot(new_fund1.Year, new_fund1.Fund_Size)
pyplot.show()
```

```python
new_fund1["Fund1's Growth"] = new_fund1.Fund_Size/new_fund1.Fund_Size.shift(1)-1
```

```python
new_fund1
```

```python
pyplot.plot(new_fund1.Year, new_fund1["Fund1's Growth"])
pyplot.show()
```

```python
sales_score = pd.DataFrame({'R^2':[-3.96,-3.65,-4.03,-5.04,-3.83,-2.41,-0.13]}, 
                           index=['Linear','Decision Tree','Random Forest','Holt Inception','Holt 10y','Holt 5y', 'Holt 3y'])
sales_score
```

```python
sales_score.plot(kind='bar', title ="R^2 Score of Sales Predictors",figsize=(15,10),legend=False, fontsize=16)
pyplot.show()
```
