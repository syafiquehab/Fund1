# Linear Regression with no lagging factor

```python
klcisales_model = linear.LinearRegression()
```

```python
# regression between FBMKLCI and Fund 1 sales
klci = fund1.loc[:,'FBMKLCI'].reshape(-1,1)
sales = fund1.Sales
```

```python
klcisales_model.fit(klci, sales)
```

```python
sales_predictions = klcisales_model.predict(klci)
```

```python
plt.scatter(fund1.FBMKLCI, fund1.Sales)
plt.plot(fund1.FBMKLCI, sales_predictions, color='red')
plt.xlabel('FBMKLCI')
plt.ylabel('Sales')
plt.title('Relationship between FBMKLCI and Sales')
```

```python
# comparing actual sales and forecasted sales
plt.scatter(fund1.Sales, sales_predictions)
plt.xlabel('Actual Sales')
plt.ylabel('Forecasted Sales')
plt.title('Actual Sales vs Forecasted Sales')
```

```python
klcisales_model.coef_
```

```python
klcisales_model.intercept_
```

```python
import sklearn.metrics as metrics
print metrics.mean_squared_error(sales, sales_predictions)
print metrics.r2_score(sales, sales_predictions)
```

```python
# regression between FBMKLCI with Fund 1 redemption
redemption = fund1.Redemption.abs() # converting redemption values to positive
klcired_model = linear.LinearRegression()
klcired_model.fit(klci, redemption)
redemption_predictions = klcired_model.predict(klci)
plt.scatter(fund1.FBMKLCI, fund1.Redemption.abs())
plt.plot(fund1.FBMKLCI, redemption_predictions, color='red')
plt.xlabel('FBMKLCI')
plt.ylabel('Redemption')
plt.title('Relationship between FBMKLCI and Redemption')
```

```python
# comparing actual redemption and forecasted redemption
plt.scatter(fund1.Redemption.abs(), redemption_predictions)
plt.xlabel('Actual Redemption')
plt.ylabel('Forecasted Redemption')
plt.title('Actual Redemption vs Forecasted Redemption')
```

```python
print klcired_model.coef_
print klcired_model.intercept_
print metrics.mean_squared_error(redemption, redemption_predictions)
print metrics.r2_score(redemption, redemption_predictions)
```

```python
# regression between sales and redemption of Fund 1
new_sales = fund1.loc[:,'Sales'].reshape(-1,1)
salesred_model = linear.LinearRegression()
salesred_model.fit(new_sales, redemption)
new_redemption_predictions = salesred_model.predict(new_sales)
plt.scatter(new_sales, redemption)
plt.plot(new_sales, new_redemption_predictions, color='red')
plt.xlabel('Sales')
plt.ylabel('Redemption')
plt.title('Relationship Between Sales and Redemption')
```

```python
# comparing actual redemption and forecasted redemption by sales pattern
plt.scatter(redemption, new_redemption_predictions)
plt.xlabel('Actual Redemption')
plt.ylabel('Forecasted Redemption')
plt.title('Actual Redemption vs Forecasted Redemption')
```

```python
print salesred_model.coef_
print salesred_model.intercept_
print metrics.mean_squared_error(redemption, new_redemption_predictions)
print metrics.r2_score(redemption, new_redemption_predictions)
```

```python
# regression between FBMKLCI and Sales with Fund 1 redemption
two_factors = fund1.loc[:,['Sales', 'FBMKLCI']]
twofactors_model = linear.LinearRegression()
twofactors_model.fit(two_factors, redemption)
twofactors_redemption_predictions = twofactors_model.predict(two_factors)
plt.scatter(redemption, twofactors_redemption_predictions)
plt.xlabel('Actual Redemption')
plt.ylabel('Forecasted Redemption')
plt.title('Actual Redemption vs Forecasted Redemption based on FBMKLCI and Sales Pattern')
```

```python
print twofactors_model.coef_
print twofactors_model.intercept_
print metrics.mean_squared_error(redemption, twofactors_redemption_predictions)
print metrics.r2_score(redemption, twofactors_redemption_predictions)
```

# Linear Regression with 1-month Lagging Factor

```python
new_fund1 = pd.read_csv('fund1.csv', thousands=',')
new_fund1
```

```python
s2 = new_fund1.Sales.drop(new_fund1.index[[292]]).reshape(-1,1) # removing latest data from sales
```

```python
r2 = new_fund1.Redemption.shift(-1).drop(new_fund1.index[[292]]).abs() # lagging 1 month redemption
```

```python
s3 = new_fund1.Sales.shift(-1).drop(new_fund1.index[[292]]) # lagging 1 month sales
```

```python
klci2 = new_fund1.FBMKLCI.drop(new_fund1.index[[292]]).reshape(-1,1) # removing latest data from FBMKLCI
```

```python
# there is a lot of outliers due to seasonality factor in January every year
```

```python
# regression between sales and lagging 1-mth redemption
lagred_model = linear.LinearRegression()
lagred_model.fit(s2, r2)
lag_redemption_predictions = lagred_model.predict(s2)
plt.scatter(s2, r2)
plt.plot(s2, lag_redemption_predictions, color='red')
plt.xlabel('Sales')
plt.ylabel('Lagging 1-Mth Redemption')
plt.title('Relationship Between Sales and Lagging 1-mth Redemption')
```

```python
# comparing lagging 1-mth actual redemption and lagging 1-mth forecasted redemption by sales pattern
plt.scatter(r2, lag_redemption_predictions)
plt.xlabel('Lagging 1-mth Actual Redemption')
plt.ylabel('Lagging 1-mth Forecasted Redemption')
plt.title('Actual Redemption vs Forecasted Redemption')
```

```python
print lagred_model.coef_
print lagred_model.intercept_
print metrics.mean_squared_error(r2, lag_redemption_predictions)
print metrics.r2_score(r2, lag_redemption_predictions)
```

```python
# regression between FBMKLCI and lagging 1-mth redemption
klcilagred_model = linear.LinearRegression()
klcilagred_model.fit(klci2, r2)
lag_redemption2_predictions = klcilagred_model.predict(klci2)
plt.scatter(klci2, r2)
plt.plot(klci2, lag_redemption2_predictions, color='red')
plt.xlabel('FBMKLCI')
plt.ylabel('Lagging 1-Mth Redemption')
plt.title('Relationship Between FBMKLCI and Lagging 1-mth Redemption')
```

```python
# comparing lagging 1-mth actual redemption and lagging 1-mth forecasted redemption by FBMKLCI pattern
plt.scatter(r2, lag_redemption2_predictions)
plt.xlabel('Lagging 1-mth Actual Redemption')
plt.ylabel('Lagging 1-mth Forecasted Redemption')
plt.title('Actual Redemption vs Forecasted Redemption')
```

```python
print klcilagred_model.coef_
print klcilagred_model.intercept_
print metrics.mean_squared_error(r2, lag_redemption2_predictions)
print metrics.r2_score(r2, lag_redemption2_predictions)
```

```python
# regression between FBMKLCI and lagging 1-mth sales
klcilagsales_model = linear.LinearRegression()
klcilagsales_model.fit(klci2, s3)
lag_sales_predictions = klcilagsales_model.predict(klci2)
plt.scatter(klci2, s3)
plt.plot(klci2, lag_sales_predictions, color='red')
plt.xlabel('FBMKLCI')
plt.ylabel('Lagging 1-Mth Sales')
plt.title('Relationship Between FBMKLCI and Lagging 1-mth Sales')
```

```python
# comparing lagging 1-mth actual sales and lagging 1-mth forecasted sales by FBMKLCI pattern
plt.scatter(s3, lag_sales_predictions)
plt.xlabel('Lagging 1-mth Actual Sales')
plt.ylabel('Lagging 1-mth Forecasted Sales')
plt.title('Actual Sales vs Forecasted Sales')
```

```python
print klcilagsales_model.coef_
print klcilagsales_model.intercept_
print metrics.mean_squared_error(s3, lag_sales_predictions)
print metrics.r2_score(s3, lag_sales_predictions)
```

```python
# regression between sales and FBMKLCI with lagging 1-mth redemption
s2_and_klci2 = new_fund1.loc[:,['FBMKLCI', 'Sales']].drop(new_fund1.index[[292]])
twofactorslag_model = linear.LinearRegression()
twofactorslag_model.fit(s2_and_klci2, r2)
twofactorslag_redemption_predictions = twofactorslag_model.predict(s2_and_klci2)
plt.scatter(r2, twofactorslag_redemption_predictions)
plt.xlabel('Lagging 1-mth Actual Redemption')
plt.ylabel('Lagging 1-mth Forecasted Redemption')
plt.title('Actual Redemption vs Forecasted Redemption based on FBMKLCI and Sales Pattern')
```

```python
print twofactorslag_model.coef_
print twofactorslag_model.intercept_
print metrics.mean_squared_error(r2, twofactorslag_redemption_predictions)
print metrics.r2_score(r2, twofactorslag_redemption_predictions)
```
